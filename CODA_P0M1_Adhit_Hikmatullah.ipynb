### Import Pandas, BeautifulSoup, Webdriver

```python
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import requests
```

### Ambil data dari Weblink

```python
import time
driver = webdriver.Chrome() ## Panggil driver-nya
nama_buku = []
penulis = []
detail_produk = []
harga_lama = []
harga_diskon = []

for halaman in range(1,7):
    weblink = f'https://www.bukabuku.com/browses/index/cid:49/c:fiksi/page:{halaman}/' ## Ambil weblink-nya
    driver.get(weblink) ## Akses weblink yang ingin dituju
    driver.find_element(By.TAG_NAME,'body').send_keys(Keys.END) ## Supaya bisa scrolling
    time.sleep(10) ## Memberikan waktu untuk mengambil data
    driver.find_element(By.TAG_NAME,'body').send_keys(Keys.HOME) ## Supaya bisa scrolling
    time.sleep(10) ## Memberikan waktu untuk mengambil data
    html = driver.page_source ## Ambil html-nya
    soup = BeautifulSoup (html, 'html.parser') ## Ekstraksi data dari html

    for kotak in soup.find_all('div', {'class':'product_list_grid'}):
        try:
            nama_buku.append(kotak.find('span', {'class':'product_list_title'}).get_text())
        except:
            nama_buku.append(None) 

        try:
            penulis.append(kotak.find('div', {'class':'product_author text_smaller'}).get_text())
        except:
            penulis.append(None)  

        try:
            detail_produk.append(kotak.find('div', {'class':'product_list_detail text_smaller'}).get_text())
        except:
            detail_produk.append(None) 

        try:
            harga_lama.append(kotak.find('span', {'class':'price_discounted'}).get_text())
        except:
            harga_lama.append(None)  

        try:
            harga_diskon.append(kotak.find('span', {'class':'price'}).get_text())
        except:
            harga_diskon.append(None)              


driver.close()
```

### Data Frame

```python
df = pd.DataFrame()
df['nama_buku'] = nama_buku
df['penulis'] = penulis
df['detail_produk'] = detail_produk 
df['harga_lama'] = harga_lama
df['harga_diskon'] = harga_diskon
```

## Transform
Pada proses transform ini dilakukan pengolahan data menggunakan Pandas.

### Menambah kolom

#### Menambah kolom `tanggal_terbit`
Proses ini bertujuan untuk menambah kolom `tanggal_terbit`. Untuk penambahan data-nya menggunakan data dari detail produk karena di kolom tersebut sudah tercantum tanggal terbit. Sisa-nya akan dilakukan penghapusan di bagian selanjutnya.

```python
df['tanggal_terbit'] = df['detail_produk'] 
```

### Menghapus Kolom

#### Menghapus kolom yang tidak diinginkan

```python
del df['Unnamed: 0']
```

```python
display(df)
```

### Menghapus kata

#### Menghapus kata pada kolom `penulis`
Proses ini bertujuan untuk menghasilkan nama penulis saja di kolom tersebut.

```python
df.penulis = df.penulis.str.replace(r'\t|\n|oleh','', regex=True)
```

#### Menghapus kata pada kolom `harga_lama` & `harga_diskon`
Proses ini bertujuan untuk menghasilkan angka saja di kolom tersebut.

```python
df.harga_lama = df.harga_lama.str.replace(r'\t|Rp','', regex=True)
df.harga_lama = df.harga_lama.str.replace('.','')
df.harga_diskon = df.harga_diskon.str.replace(r'\t|Rp','', regex=True)
df.harga_diskon = df.harga_diskon.str.replace('.','')
```

#### Menghapus kata pada kolom `detail_produk`
Proses ini bertujuan untuk menghasilkan data Hard Cover atau Soft Cover saja di kolom tersebut.

```python
df.detail_produk = df.detail_produk.str.replace(r'\n|Agustus|Juli|Juni|Mei|April|Maret|Februari|2025','', regex=True)
df.detail_produk = df.detail_produk.str.replace(',','')
```

#### Menghapus kata pada kolom `tanggal_terbit`
Proses ini bertujuan untuk menghasilkan data tanggal terbit saja di kolom tersebut.

```python
df.tanggal_terbit = df.tanggal_terbit.str.replace(r'\n|Soft|Hard|Cover','', regex=True)
df.tanggal_terbit = df.tanggal_terbit.str.replace(',','')
```

### Filter data yang masih bernilai `none` 

#### Mencari data buku yang masih bernilai `none`

#### Data yang berisikan `none` dihapuskan

```python
df.dropna(inplace=True)
```

### Mengubah Format Data

#### Mengubah format data
Data type tiap kolom diubah, seperti `nama_buku`, `penulis`, `detail_produk`, dan `tanggal_terbit` diubah menjadi `str`. Sedangkan untuk `harga_lama` dan `harga_diskon` diubah menjadi integer.

```python
df.nama_buku = df.nama_buku.astype('str')
df.penulis = df.penulis.astype('str')
df.detail_produk = df.detail_produk.astype('str')
df.harga_lama = df.harga_lama.astype('int')
df.harga_diskon = df.harga_diskon.astype('int')
df.tanggal_terbit = df.tanggal_terbit.astype('str')
```
